**Ejercicio 1**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Datos e intervalos (amplitud 2)
# -----------------------------
datos = [2, 3, 5, 1, 6, 7, 8, 0, 8, 9, 6, 4, 4, 3, 2, 6,
         3, 3, 2, 1, 1, 5, 6, 7, 4, 3, 2, 5, 4, 6]
amplitud = 2
bins = np.arange(min(datos), max(datos) + amplitud + 1, amplitud)  # +1 para cubrir borde superior

# -----------------------------
# Tabla de frecuencias completa
# -----------------------------
fa, _ = np.histogram(datos, bins=bins)
fr = fa / len(datos)
fac = np.cumsum(fa)
frac = np.cumsum(fr)

intervalos = pd.IntervalIndex.from_breaks(bins, closed='left')
tabla = pd.DataFrame({
    "Intervalo": intervalos.astype(str),
    "Frecuencia Absoluta": fa,
    "Frecuencia Relativa (%)": np.round(fr * 100, 2),
    "Frecuencia Acumulada": fac,
    "Frecuencia Rel. Acum. (%)": np.round(frac * 100, 2),
})
print("TABLA DE FRECUENCIAS:\n")
print(tabla.to_string(index=False))

# -----------------------------
# Histograma con barras coloreadas (una por clase)
# -----------------------------
centros = (bins[:-1] + bins[1:]) / 2
ancho = amplitud * 0.9  # leve espacio entre barras
cmap = plt.cm.get_cmap('tab20', len(fa))
colores = [cmap(i) for i in range(len(fa))]

plt.figure(figsize=(9, 5))
plt.bar(centros, fa, width=ancho, color=colores, edgecolor="black", label="Frecuencia por intervalo")

# Ejes, títulos y rejilla
plt.title("Histograma (Intervalos de Amplitud 2)", fontsize=14, fontweight="bold")
plt.xlabel("Intervalos", fontsize=12)
plt.ylabel("Frecuencia Absoluta", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Etiquetas de intervalos en el eje X
plt.xticks(centros, [str(iv) for iv in intervalos])

# Valores encima de cada barra
for x, y in zip(centros, fa):
    plt.text(x, y + 0.2, str(y), ha='center', va='bottom', fontsize=10)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Datos
tipos_auto = ["SUV", "Sedán", "Camioneta", "Deportivo", "SUV", "SUV", "Sedán",
              "Camioneta", "Sedán", "SUV", "Deportivo", "Sedán", "Camioneta"]

# Frecuencias absolutas y relativas
valores, fa = np.unique(tipos_auto, return_counts=True)
fr = np.round((fa / len(tipos_auto)) * 100, 2)

# Tabla
tabla = pd.DataFrame({
    "Tipo de Auto": valores,
    "Frecuencia Absoluta": fa,
    "Frecuencia Relativa (%)": fr
})
print("TABLA DE TIPOS DE AUTO:\n")
print(tabla.to_string(index=False))

# Colores
cmap = plt.cm.get_cmap('tab20', len(valores))
colores = [cmap(i) for i in range(len(valores))]

# Diagrama de sectores
plt.figure(figsize=(7, 7))
plt.pie(fr, labels=valores, autopct="%1.1f%%", startangle=90, colors=colores)
plt.title("Distribución de Tipos de Autos", fontsize=14, fontweight="bold")
plt.legend(title="Tipo", loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
