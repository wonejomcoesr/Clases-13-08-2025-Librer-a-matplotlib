**Ejercicio_01**

Preguntar a 5 estudiantes sus edades, hallar la media y promedio; finalmente utilizando las barras mostrar la media y el promedio

```python
import matplotlib.pyplot as plt
import numpy as np

edades = []

# Pedir edades a 5 estudiantes
for i in range(5):
    edad = int(input(f"Ingresa la edad del estudiante {i+1}: "))
    edades.append(edad)

# Calcular media y mediana
media = np.mean(edades)
mediana = np.median(edades)

print(f"Edades: {edades}")
print(f"Media (promedio): {media:.2f}")
print(f"Mediana: {mediana}")

# Crear gráfico de barras
fig, ax = plt.subplots()

# Barras para las edades individuales
ax.bar(range(1, 6), edades, label='Edades estudiantes')

# Barras para la media y la mediana (en posiciones 6 y 7 para que no se mezclen)
ax.bar(6, media, color='orange', label='Media')
ax.bar(7, mediana, color='green', label='Mediana')

# Etiquetas y títulos
ax.set_xlabel('Estudiantes / Estadísticas')
ax.set_ylabel('Edad')
ax.set_title('Edades, Media y Mediana de Estudiantes')
ax.set_xticks(range(1, 8))
ax.set_xticklabels([f'Est {i}' for i in range(1, 6)] + ['Media', 'Mediana'])

ax.legend()
plt.show()
```

**Ejercicio_02: Diagrama pastel**

Sacar 2 graficos:

Del ejercicio anterior preguntar el peso de los studiantes en kg y sacar, la media y la desviacion standar.

Sacar el porcentaje de varones y porcentaje de mujeres del salon

```python
import matplotlib.pyplot as plt
import numpy as np

edades = []
pesos = []

# Pedir edad y peso de 5 estudiantes
for i in range(5):
    edad = int(input(f"Ingresa la edad del estudiante {i+1}: "))
    edades.append(edad)

    peso = float(input(f"Ingresa el peso (kg) del estudiante {i+1}: "))
    pesos.append(peso)

# Calcular media y desviación estándar para edades y pesos
media_edad = np.mean(edades)
desv_edad = np.std(edades)

media_peso = np.mean(pesos)
desv_peso = np.std(pesos)

print(f"\nEdades ingresadas: {edades}")
print(f"Pesos ingresados: {pesos}")
print(f"Media de las edades: {media_edad:.2f} años")
print(f"Desviación estándar de las edades: {desv_edad:.2f} años")
print(f"Media del peso: {media_peso:.2f} kg")
print(f"Desviación estándar del peso: {desv_peso:.2f} kg")

# Preguntar cantidad de varones y mujeres en el salón
varones = int(input("\n¿Cuántos varones hay en el salón? "))
mujeres = int(input("¿Cuántas mujeres hay en el salón? "))
total = varones + mujeres

porc_varones = (varones / total) * 100 if total > 0 else 0
porc_mujeres = (mujeres / total) * 100 if total > 0 else 0

print(f"\nTotal de alumnos: {total}")
print(f"Porcentaje de varones: {porc_varones:.1f}%")
print(f"Porcentaje de mujeres: {porc_mujeres:.1f}%")

# Gráfico 1: diagrama pastel para media y desviación estándar de las edades
fig1, ax1 = plt.subplots()
labels_edad = ['Media edad', 'Desviación estándar edad']
sizes_edad = [media_edad, desv_edad]
colors_edad = ['gold', 'lightcoral']
explode_edad = (0.1, 0)

ax1.pie(sizes_edad, labels=labels_edad, colors=colors_edad, autopct='%1.2f', startangle=90, explode=explode_edad)
ax1.set_title('Media y Desviación Estándar de las Edades')
ax1.axis('equal')

# Gráfico 2: diagrama pastel para media y desviación estándar del peso
fig2, ax2 = plt.subplots()
labels_peso = ['Media peso', 'Desviación estándar peso']
sizes_peso = [media_peso, desv_peso]
colors_peso = ['skyblue', 'lightgreen']
explode_peso = (0.1, 0)

ax2.pie(sizes_peso, labels=labels_peso, colors=colors_peso, autopct='%1.2f', startangle=90, explode=explode_peso)
ax2.set_title('Media y Desviación Estándar del Peso (kg)')
ax2.axis('equal')

# Gráfico 3: diagrama pastel para porcentaje varones y mujeres
fig3, ax3 = plt.subplots()
labels_sexo = ['Varones', 'Mujeres']
sizes_sexo = [porc_varones, porc_mujeres]
colors_sexo = ['deepskyblue', 'lightpink']
explode_sexo = (0.1, 0)

ax3.pie(sizes_sexo, labels=labels_sexo, colors=colors_sexo, autopct='%1.1f%%', startangle=90, explode=explode_sexo)
ax3.set_title('Porcentaje de Varones y Mujeres en el Salón')
ax3.axis('equal')

plt.show()
```

**Diagramas de contorno**

contourf(x, y, z): Dibuja un diagrama de contorno con las curvas de nivel de la superficie dada por los puntos con las coordenadas de las listas x, y y z en los ejes X, Y y Z respectivamente.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
x, y = np.meshgrid(x, y)
z = np.sqrt(x**2 + 2*y**2)
ax.contourf(x, y, z)
plt.show()
```

**Colores**

Para cambiar el color de los objetos se utiliza el parámetro color = nombre-color, donde nombre-color es una cadena con el nombre del color de entre los colores disponibles.



```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
dias = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
temperaturas = {'Madrid':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'Barcelona':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25]}
ax.plot(dias, temperaturas['Madrid'], color = 'tab:purple')
ax.plot(dias, temperaturas['Barcelona'], color = 'tab:green')
plt.show()
```

**Títulos**

Para añadir un título principal al gráfico se utiliza el siguiente método:

ax.set_title(titulo, loc=alineacion, fontdict=fuente) : Añade un título con el contenido de la cadena titulo a los ejes ax. El parámetro loc indica la alineación del título, que puede ser 'left' (izquierda), 'center' (centro) o 'right' (derecha), y el parámetro fontdict indica mediante un diccionario las características de la fuente (la el tamaño fontisize, el grosor fontweight o el color color).

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
dias = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
temperaturas = {'Madrid':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'Barcelona':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25]}
ax.plot(dias, temperaturas['Madrid'])
ax.plot(dias, temperaturas['Barcelona'])
ax.set_title('Evolución de la temperatura diaria', loc = "left", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
plt.show()
```

**Ejes**

Para cambiar el aspecto de los ejes se suelen utilizar los siguientes métodos:

ax.set_xlabel(titulo) : Añade un título con el contenido de la cadena titulo al eje x de ax. Se puede personalizar la alineación y la fuente con los mismos parámetros que para el título principal.
ax.set_ylabel(titulo) : Añade un título con el contenido de la cadena titulo al eje y de ax. Se puede personalizar la alineación y la fuente con los mismos parámetros que para el título principal.
ax.set_xlim([limite-inferior, limite-superior]) : Establece los límites que se muestran en el eje x de ax.
ax.set_ylim([limite-inferior, limite-superior]) : Establece los límites que se muestran en el eje y de ax.
ax.set_xticks(marcas) : Dibuja marcas en el eje x de ax en las posiciones indicadas en la lista marcas.
ax.set_yticks(marcas) : Dibuja marcas en el eje y de ax en las posiciones indicadas en la lista marcas.
ax.set_xscale(escala) : Establece la escala del eje x de ax, donde el parámetro escala puede ser 'linear' (lineal) o 'log' (logarítmica).
ax.set_yscale(escala) : Establece la escala del eje y de ax, donde el parámetro escala puede ser 'linear' (lineal) o 'log' (logarítmica).

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
dias = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
temperaturas = {'Madrid':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'Barcelona':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25]}
ax.plot(dias, temperaturas['Madrid'])
ax.plot(dias, temperaturas['Barcelona'])
ax.set_xlabel("Días", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
ax.set_ylabel("Temperatura ºC")
ax.set_ylim([20,35])
ax.set_yticks(range(20, 35))
plt.show()
```

**Leyenda**

Para añadir una leyenda a un gráfico se utiliza el siguiente método:

ax.legend(leyendas, loc = posición) : Dibuja un leyenda en los ejes ax con los nombres indicados en la lista leyendas. El parámetro loc indica la posición en la que se dibuja la leyenda y puede ser 'upper left' (arriba izquierda), 'upper center' (arriba centro), 'upper right' (arriba derecha), 'center left' (centro izquierda), 'center' (centro), 'center right' (centro derecha), 'lower left' (abajo izquierda), 'lower center' (abajo centro), 'lower right' (abajo derecha). Se puede omitir la lista leyendas si se indica la leyenda de cada serie en la función que la dibuja mediante el parámetro label.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
dias = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
temperaturas = {'Madrid':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'Barcelona':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25]}
ax.plot(dias, temperaturas['Madrid'], label = 'Madrid')
ax.plot(dias, temperaturas['Barcelona'], label = 'Barcelona')
ax.legend(loc = 'upper right')
plt.show()
```
